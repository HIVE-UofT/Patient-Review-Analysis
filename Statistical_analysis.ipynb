```python
import pandas as pd
import numpy as np

# Filter dataset to include only positive and negative sentiments, exclude neutral/mixed
df_filtered = df[df['ProcessedValence'].isin(['Positive', 'Negative'])].copy()

# Exclude unknown units
df_filtered = df_filtered[df_filtered['ProcessedUnit'] != 'Unknown'].copy()

# Create binary outcome: 1 for Positive, 0 for Negative
df_filtered['SentimentBinary'] = (df_filtered['ProcessedValence'] == 'Positive').astype(int)

# Convert CovidPeriod to integer (0 = Pre-COVID, 1 = During-COVID)
df_filtered['CovidPeriod'] = df_filtered['CovidPeriod'].astype(int)

# Center the Year (optional, for interpretability of intercepts in models)
df_filtered['YearCentered'] = df_filtered['Year'] - 2019

# Quick check of the filtered data
print(df_filtered['ProcessedValence'].value_counts())  # Should show only Positive and Negative counts
print(df_filtered['ProcessedUnit'].value_counts().head(5))  # Top 5 units after filtering (Unknown removed)
```

```python
import scipy.stats as stats

# Create a contingency table of Year vs Sentiment (Positive=1/Negative=0)
year_sent_ct = pd.crosstab(df_filtered['Year'], df_filtered['SentimentBinary'])
print(year_sent_ct)

# Chi-square test for association between Year and sentiment
chi2, p, dof, exp = stats.chi2_contingency(year_sent_ct)
print(f"Chi-square test: chi2={chi2:.2f}, df={dof}, p-value={p:.4f}")

# Calculate Cramer's V (phi for 2-category vs multi-year)
N = year_sent_ct.values.sum()
phi = np.sqrt(chi2 / (N * min(year_sent_ct.shape[0]-1, year_sent_ct.shape[1]-1)))
print(f"Effect size (Cramer's V): {phi:.3f}")

# Compute proportion of positive reviews each year
prop_pos_year = year_sent_ct[1] / (year_sent_ct[0] + year_sent_ct[1])
print("Proportion of positive sentiment by year:")
print(prop_pos_year.round(3))
```

```python
from statsmodels.stats.proportion import proportions_ztest

# Define the year groups for pre-COVID and during-COVID
pre_years = [2017, 2018, 2019]
covid_years = [2020, 2021, 2022]

# Filter data for the two groups
df_pre = df_filtered[df_filtered['Year'].isin(pre_years)]
df_covid = df_filtered[df_filtered['Year'].isin(covid_years)]

# Count positive and negative reviews in each period
pos_pre = df_pre['SentimentBinary'].sum()
neg_pre = len(df_pre) - pos_pre
pos_covid = df_covid['SentimentBinary'].sum()
neg_covid = len(df_covid) - pos_covid

# Two-proportion z-test for difference in positive sentiment proportion
count = np.array([pos_pre, pos_covid])
nobs = np.array([pos_pre + neg_pre, pos_covid + neg_covid])
z_stat, p_val = proportions_ztest(count, nobs)

# Calculate proportions for reporting
prop_pre = pos_pre / (pos_pre + neg_pre)
prop_covid = pos_covid / (pos_covid + neg_covid)
print(f"Positive sentiment proportion pre-COVID (2017-2019): {prop_pre:.3f}")
print(f"Positive sentiment proportion during COVID (2020-2022): {prop_covid:.3f}")
print(f"Two-proportion z-test: Z = {z_stat:.3f}, p-value = {p_val:.4f}")
```

```python
import statsmodels.formula.api as smf

# Treat ProcessedUnit as categorical and set a reference category (baseline) for interpretability
df_filtered['ProcessedUnit'] = df_filtered['ProcessedUnit'].astype('category')
# Set 'Inpatient Ward' as baseline unit if present (common unit with large count)
if 'Inpatient Ward' in df_filtered['ProcessedUnit'].cat.categories:
    cats = ['Inpatient Ward'] + [u for u in df_filtered['ProcessedUnit'].cat.categories if u != 'Inpatient Ward']
    df_filtered['ProcessedUnit'] = df_filtered['ProcessedUnit'].cat.reorder_categories(cats, ordered=False)

# Logistic regression model: Sentiment ~ Year + Unit + Year*Unit
model1 = smf.logit('SentimentBinary ~ YearCentered * C(ProcessedUnit)', data=df_filtered).fit()
print(model1.summary())

# Test overall significance of the interaction (Year:Unit) via likelihood ratio test
model1_noInt = smf.logit('SentimentBinary ~ YearCentered + C(ProcessedUnit)', data=df_filtered).fit(disp=0)
```

```python
from scipy.stats import chi2

# Manual likelihood ratio test
llf_full = model1.llf       # log-likelihood of full model
llf_reduced = model1_noInt.llf  # log-likelihood of reduced model
lr_stat = 2 * (llf_full - llf_reduced)
lr_df = model1.df_model - model1_noInt.df_model
lr_p = chi2.sf(lr_stat, lr_df)

print(f"Likelihood-ratio test for interaction: chi2 = {lr_stat:.2f}, df = {lr_df}, p = {lr_p:.4f}")
```

```python
import statsmodels.api as sm
from scipy.stats import chi2

# One-hot encode theme categories and interaction terms
df_codes = df_codes.copy()
df_codes['Intercept'] = 1
theme_dummies = pd.get_dummies(df_codes['ThemeCode'], prefix="Theme", drop_first=True)
X_base = pd.concat([df_codes[['Intercept', 'CovidPeriod']], theme_dummies], axis=1)

# Add interaction terms: Theme × CovidPeriod
for col in theme_dummies.columns:
    X_base[f"{col}_x_Covid"] = theme_dummies[col] * df_codes['CovidPeriod']

y = df_codes['SentimentBinary'].astype(int)

X_base = X_base.astype(float)
y = y.astype(int)

# Fit penalized GLM (L2 ridge) to stabilize
model_glm = sm.GLM(y, X_base, family=sm.families.Binomial())
res_pen = model_glm.fit_regularized(method='elastic_net', L1_wt=0.0, alpha=1.0)

# Refit without penalty to get inference stats
res_full = model_glm.fit(start_params=res_pen.params)

# Fit reduced model without interaction
X_reduced = X_base.drop(columns=[c for c in X_base.columns if '_x_Covid' in c])
X_reduced = X_reduced.astype(float)
model_glm_reduced = sm.GLM(y, X_reduced, family=sm.families.Binomial())
res_reduced = model_glm_reduced.fit()

# Likelihood ratio test
llf_full = res_full.llf
llf_reduced = res_reduced.llf
lr_stat = 2 * (llf_full - llf_reduced)
lr_df = res_full.df_model - res_reduced.df_model
lr_p = chi2.sf(lr_stat, lr_df)

summary_text = res_full.summary2().as_text()
lr_result = f"LR test for CovidPeriod × ThemeCode: chi2 = {lr_stat:.2f}, df = {lr_df}, p = {lr_p:.4f}"
summary_text, lr_result
```

```python
import logging
# 1. Load model
nlp = spacy.load("en_ner_bc5cdr_md")

# (Optional) add a SpanCategorizer if you train one
# nlp.add_pipe("my_spancat", after="ner")

# Set up logging
logging.basicConfig(
    filename="extraction_errors.log",
    level=logging.ERROR,
    format="%(asctime)s %(levelname)s:%(message)s"
)
```

```python
# Global accumulators
entities_count = {}
unique_entities = set()

def extract_drug_disease(comment, threshold=0.5):
    text = str(comment).lower()
    mentions = []
    try:
        doc = nlp(text)
    except Exception as e:
        logging.error(f"Failed to process doc: {e} -- {text[:50]}")
        return mentions

    # 2. Gather raw spans + optional score
    spans = []
    for ent in doc.ents:
        # Example: if using spancat, you'd have ent._.score
        score = getattr(ent._, "score", None)  # None if no spancat
        # Fallback: treat no-score as high-confidence
        score = score if score is not None else 1.0
        if score < threshold:
            continue

        # 3. Optional post-filter: check vocab vector_norm
        lex = nlp.vocab[ent.text]
        if not lex.has_vector or lex.vector_norm == 0:
            continue

        spans.append(ent.text)

    # 4. Count overlapping mentions via lookahead
    for span_text in spans:
        pattern = re.compile(rf"(?=({re.escape(span_text)}))")
        for m in pattern.finditer(text):
            s = m.group(1)
            mentions.append((s, None))  # or carry ent.label_ if needed
            entities_count[s] = entities_count.get(s, 0) + 1
            unique_entities.add(s)

    return mentions
```
