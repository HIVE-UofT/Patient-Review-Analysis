```python
import spacy
import scispacy
from scispacy.linking import EntityLinker
import pandas as pd
import re
from tqdm import tqdm
```

```python
import logging
# 1. Load model
nlp = spacy.load("en_ner_bc5cdr_md")

# (Optional) add a SpanCategorizer if you train one
# nlp.add_pipe("my_spancat", after="ner")

# Set up logging
logging.basicConfig(
    filename="extraction_errors.log",
    level=logging.ERROR,
    format="%(asctime)s %(levelname)s:%(message)s"
)
```

```python
# Global accumulators
entities_count = {}
unique_entities = set()

def extract_drug_disease(comment, threshold=0.5):
    text = str(comment).lower()
    mentions = []
    try:
        doc = nlp(text)
    except Exception as e:
        logging.error(f"Failed to process doc: {e} -- {text[:50]}")
        return mentions

    # 2. Gather raw spans + optional score
    spans = []
    for ent in doc.ents:
        # Example: if using spancat, you'd have ent._.score
        score = getattr(ent._, "score", None)  # None if no spancat
        # Fallback: treat no-score as high-confidence
        score = score if score is not None else 1.0
        if score < threshold:
            continue

        # 3. Optional post-filter: check vocab vector_norm
        lex = nlp.vocab[ent.text]
        if not lex.has_vector or lex.vector_norm == 0:
            continue

        spans.append(ent.text)

    # 4. Count overlapping mentions via lookahead
    for span_text in spans:
        pattern = re.compile(rf"(?=({re.escape(span_text)}))")
        for m in pattern.finditer(text):
            s = m.group(1)
            mentions.append((s, None))  # or carry ent.label_ if needed
            entities_count[s] = entities_count.get(s, 0) + 1
            unique_entities.add(s)

    return mentions
```

```python
tqdm.pandas()
df['Entities'] = df['Comment'].progress_apply(extract_drug_disease)
```

```python
# 6. Inspect top entities
entity_df = (
    pd.DataFrame.from_dict(entities_count, orient='index', columns=['Count'])
      .sort_values('Count', ascending=False)
)
print(entity_df.head(20))
```

```python
import ast

# 1. Make sure your Entities column is real lists of (text,label) tuples
def ensure_list(ents):
    if isinstance(ents, str):
        # convert from the literal “[('fever', 'DISEASE'), …]” to an actual list
        try:
            return ast.literal_eval(ents)
        except Exception:
            return []
    elif isinstance(ents, list):
        return ents
    else:
        return []

df['Entities'] = df['Entities'].apply(ensure_list)

# 2. Explode into one row per entity
df_exp = (
    df
    .explode('Entities')                  # now each row has a single tuple or NaN
    .dropna(subset=['Entities'])          # drop any empty
)

# 3. Split that tuple into two columns: EntityText and Label
df_exp[['EntityText','Label']] = pd.DataFrame(
    df_exp['Entities'].tolist(),
    index=df_exp.index
)

# 4. Count per sentiment (“ProcessedValence” should be your sentiment column)
counts = (
    df_exp
    .groupby(['ProcessedValence','EntityText'])
    .size()
    .reset_index(name='Count')
)

# 5. Get top-10 for each class
top_pos = (
    counts[counts['ProcessedValence']=='Positive']
    .nlargest(10, 'Count')
)
top_neg = (
    counts[counts['ProcessedValence']=='Negative']
    .nlargest(10, 'Count')
)

print("Top 10 entities in positive reviews:\n", top_pos[['EntityText','Count']])
print("\nTop 10 entities in negative reviews:\n", top_neg[['EntityText','Count']])
```
