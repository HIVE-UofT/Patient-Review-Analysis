```python
import re

def process_comment(row):
    comment = str(row['Comment'])
    flag = False
    lower = re.findall(r'\w+', comment.lower())
    text = ' '.join(lower)
    tokens = word_tokenize(text)
    lemmatized_tokens = [lemmatizer.lemmatize(token) for token in tokens]
    for term in covid_related_words:
        if term in lemmatized_tokens:
            flag= True
    text = ' '.join(lemmatized_tokens)
    return flag
```

```python
import re

pattern = r'x{3,}'

# Function to remove substrings matching the pattern
def remove_substrings(s):
    return re.sub(pattern, '', s)
```

```python
import ast


def getEntityDict(CovidPeriod):
    positive_dict = {}
    negative_dict = {}

    for index, row in df.iterrows():
        if row['CovidPeriod'] == CovidPeriod:
            if  row['RefinedValence'] == "Positive":
                ent_tuples = ast.literal_eval(row['Entities'])
                for ent in ent_tuples:
                    entity, label = ent
                   
                    if ((entity not in positive_dict.keys()) and (not entity.startswith('xx'))):
                        positive_dict[entity] = 1
                    elif entity in positive_dict.keys():
                        positive_dict[entity] += 1
            elif row['RefinedValence'] == "Negative":
                ent_tuples = ast.literal_eval(row['Entities'])
                for ent in ent_tuples:
                    entity, label = ent

                    if ((entity not in negative_dict.keys()) and (not entity.startswith('xx'))):
                        negative_dict[entity] = 1
                    elif entity in negative_dict.keys():
                        negative_dict[entity] += 1
        
    cleaned_positive_dict = {remove_substrings(k): v for k, v in positive_dict.items()}
    sorted_positive_dict = dict(sorted(cleaned_positive_dict.items(), key=lambda item: item[1]))

    cleaned_negative_dict = {remove_substrings(k): v for k, v in negative_dict.items()}
    sorted_negative_dict = dict(sorted(cleaned_negative_dict.items(), key=lambda item: item[1]))

    return sorted_positive_dict, sorted_negative_dict
```

```python
def count_vaccine():
    pos_cnt = 0
    neg_cnt = 0
    for index, row in df.iterrows():
        if row['CovidPeriod']:
            comment = str(row['Comment'])
            lower = re.findall(r'\w+', comment.lower())
            text = ' '.join(lower)
            tokens = word_tokenize(text)
            lemmatized_tokens = [lemmatizer.lemmatize(token) for token in tokens]
            for term in vaccine_related_words:
                if term in lemmatized_tokens:
                    if row['RefinedValence'] == "Positive/Negative" or row['RefinedValence'] == "Positive":
                        pos_cnt += 1
                    elif row['RefinedValence'] == "Positive/Negative" or row['RefinedValence'] == "Negative":
                        neg_cnt += 1

    return pos_cnt, neg_cnt
```

```python
import ast

def generateCometChartData():
    
    rev_dicts = {}

    for index, row in df.iterrows():
        if str(row['Refined_Type']) == "nan":
            continue
        if str(row['RefinedCode']) == "nan":
            continue
                
        ref_type = str(row['Refined_Type'])
        codes = ast.literal_eval(str(row['RefinedCode']))

        for code in codes:
            
            if str(code) == "unknown" or str(code) == "met- access/coord of care":
                continue
            period = 'COVID' if row['CovidPeriod'] else 'NonCOVID'
            if f'{ref_type} + {code} + {period}' not in rev_dicts.keys():
                rev_dicts[f'{ref_type} + {code} + {period}'] = {'neg_cnt': 0, 'pos_cnt': 0, 'cnt': 0, 'neg_perc': 0, 'pos_perc': 0}
            
            rev_dicts[f'{ref_type} + {code} + {period}']['cnt'] += 1
            if row['RefinedValence'] == "Negative":
                rev_dicts[f'{ref_type} + {code} + {period}']['neg_cnt'] += 1
                rev_dicts[f'{ref_type} + {code} + {period}']['neg_perc'] = round((rev_dicts[f'{ref_type} + {code} + {period}']['neg_cnt'] / rev_dicts[f'{ref_type} + {code} + {period}']['cnt'])*100 , 2)
            
            elif row['RefinedValence'] == "Positive":
                rev_dicts[f'{ref_type} + {code} + {period}']['pos_cnt'] += 1
                rev_dicts[f'{ref_type} + {code} + {period}']['pos_perc'] = round((rev_dicts[f'{ref_type} + {code} + {period}']['pos_cnt'] / rev_dicts[f'{ref_type} + {code} + {period}']['cnt'])*100 , 2)
            
            
            

    return rev_dicts
```

```python
def getUnitDict():
    negative_dict = {}
    positive_dict = {}
    neutral_dict = {}
    posneg_dict = {}

    
    for index, row in filtered_df.iterrows():
        unit =  str(row['ProcessedUnit'])

        if row['RefinedValence'] == "Negative":
                if unit not in negative_dict.keys():
                    negative_dict[unit] = 1
                elif unit in negative_dict.keys():
                    negative_dict[unit] += 1
        elif row['RefinedValence'] == "Positive":
                if unit not in positive_dict.keys():
                    positive_dict[unit] = 1
                elif unit in positive_dict.keys():
                    positive_dict[unit] += 1
        elif row['RefinedValence'] == "Neutral":
                if unit not in neutral_dict.keys():
                    neutral_dict[unit] = 1
                elif unit in neutral_dict.keys():
                    neutral_dict[unit] += 1
        else:
                if unit not in posneg_dict.keys():
                    posneg_dict[unit] = 1
                elif unit in posneg_dict.keys():
                    posneg_dict[unit] += 1


    return negative_dict, positive_dict, neutral_dict, posneg_dict
```
